```python
import pandas as pd
```

```python
articles = pd.read_excel('articles.xlsx',names=['In','Text'])
articles.head()
```

```python
stop_words = pd.read_csv('stopwords.csv',names=['word'])
stop_words.head()
```

```python
import re
def preparing_data(some_column):
    some_column = some_column.apply(lambda x: x.lower())
    some_column = some_column.str.replace("[^\w\s]", "", regex=True)
    some_column = some_column.str.replace("‰Ûª", "''")
    some_column = some_column.apply(lambda x: re.sub(r'[^\w\s]+|[\d]+', r'',x).strip())
    some_column = some_column.apply(lambda x: ' '.join([item for item in x.split() if item not in stop_words['word'].to_list()]))
    return some_column
    
articles['Text'] = preparing_data(articles['Text'])
articles.head()
```

```python
from nltk.collocations import *
import nltk
def pmi(column):
    common = []
    for text in column.to_list():
        finder = BigramCollocationFinder.from_words(text.split())
        bigram_measures = nltk.collocations.BigramAssocMeasures()
        pmi = finder.score_ngrams(bigram_measures.pmi)
        common.extend(pmi)
    df = pd.DataFrame(common, columns=['Pair','PMI'])
    return df

new_data = pmi(articles['Text'])
new_data.head()
```

```python
from sklearn import preprocessing
import numpy as np
x_array = np.array(new_data['PMI'])
new_data['NPMI'] = pd.Series(*preprocessing.normalize([x_array]))
new_data.head()
```

```python
new_data = new_data.drop('PMI', axis='columns').iloc[:50]
```

```python
for pair, npmi in new_data.values:
    print(f"|{pair[0]} {pair[1]} |{npmi}|")
```
